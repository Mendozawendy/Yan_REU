```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

import pandas as pd
df = pd.read_csv('/Users/wendymendoza/Desktop/DATA POINTS.csv')

# Create subplots with shared x-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Loop through your columns and add traces
for col in df.columns[1:]:  # Skip the first column ('J.D.-2400000')
    fig.add_trace(go.Scatter(x=df['J.D.-2400000'], y=df[col], mode='lines+markers', name=col), secondary_y=True)

# Update layout and formatting
fig.update_layout(
    title='Overall Multiple Flux of Astronomical Datasets',
    xaxis_title='J.D.-2400000',
    yaxis_title='Relative Flux',
    legend=dict(x=0.05, y=1),  # Adjust the position of the legend
    font=dict(family="Arial", size=12),
    margin=dict(t=80),  # Add some space at the top for the title
    showlegend=True,
    plot_bgcolor='rgb(240, 240, 240)',  # Set background color
    paper_bgcolor='rgb(255, 255, 255)',  # Set plot area background color
)

# Customize marker colors and sizes
marker_colors = px.colors.qualitative.Plotly
for i, trace in enumerate(fig.data):
    trace.marker.color = marker_colors[i % len(marker_colors)]
    trace.marker.size = 8

# Show the figure
fig.show()
```
